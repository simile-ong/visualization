# seaborn

```python
# loading packages

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
# creating basic plot

sns.lineplot(data=tips,       # choosing dataset
             x="total_bill",    # defining x variable
             y = "tip")        # defining y variable

# faint lines represent less frequent tip amounts
```

```python
# using premade styles to alter appearance of our plot

sns.set_style("darkgrid")
sns.lineplot(data=tips,
             x="total_bill",
             y = "tip")

# using default to reset style
```

```python
# adding title and axis labels

tipgraph = sns.lineplot(data=tips, x="total_bill", y = "tip")

tipgraph.set(title = "Tips vs. Total Bills",
             xlabel = "Total Bill ($)",
             ylabel = "Tip Amount ($)")
```

```python
# modifying aesthetic  -- changing figure size

fig = plt.subplots(figsize = (10,3))
tipgraph = sns.lineplot(data=tips, x="total_bill", y = "tip")

tipgraph.set(title = "Tips vs. Total Bills",
             xlabel = "Total Bill ($)",
             ylabel = "Tip Amount ($)")
```

```python
tipgraph = sns.lineplot(data=tips, x = "total_bill", y = "tip",
                        color="hotpink", linestyle = "--",
                        linewidth = 3,
                        marker = "o",
                        markerfacecolor = "indigo")

tipgraph.set(title = "Tips vs. Total Bills",
             xlabel = "Total Bill ($)",
             ylabel = "Tip Amount ($)")
```

Seaborn makes it very easy to represent multiple variables with different visual elements of our graph

```python
# adding more variables

tipgraph = sns.scatterplot(data=tips, x="total_bill", y="tip",
                           style = "time",
                           hue = "day",
                           palette = ["purple", "hotpink", "deepskyblue", "yellowgreen"])

tipgraph.set(title= "Tips vs. Total Bill",
             xlabel="Total Bill ($)",
             ylabel="Tip Amount ($)")
```

```python
# pairplot to compare variables - helps us explore data

sns.pairplot(data = tips, hue = "day")
```

```python
# relplot to explore specific levels within variable

daysplot = sns.relplot(data=tips, x="total_bill", y="tip",
                      hue = "sex",
                      col = "day",
                      kind = "scatter",
                      col_wrap=2)
```

```python
# loading libraries

import plotly.graph_objects as go
from wordcloud import WordCloud
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
```

```python
x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
# making plot

graph = go.Figure()
graph.add_trace(go.Bar(x=x1, y=y1))
```

```python
graph.update_layout(
    title="Pirate Scores",
    xaxis_title="Pirates",
    yaxis_title="Score"
)
graph.show()
```

## plotly features

If we hover over the bars on our plot, we get automatic data labels

If we click and drag over a portion of our plot, we can zoom in
Double click to zoom back out

We can click and drag on our axis to scroll

In the upper-right menu:
    Save as PNG, Pan/zoom, Box/Lasso select

We can save plotly graphs as HTML files and embed in webpage -- graph.write_html("[filepath]/pirategraph.html")

```python
# customizing plotly graphs

graph = go.Figure()
graph.add_trace(go.Scatter(x=x1, y=y1, mode = "markers",    # we want points for scatter points
                           marker=dict(
                               size=15,              # point size
                               color="hotpink",     # point color
                               opacity=1,           # point transparency/ alpha
                               line=dict(width=5, color="purple")   # point outline
                           )))

graph.update_layout(
    title = "Interactive Pirate Plot",
    xaxis_title = "Pirates",
    yaxis_title = "Scores",
    width = 500, height = 500
)
```

```python
# note: already imported wordcloud above
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",on_bad_lines="skip")
df
```

```python
# making wordclouds

# join all text from each row from our quote column into a string
text = " ".join(each for each in df.quote)

# generate wordcloud image
wordcloud = WordCloud(background_color="white", colormap="inferno").generate(text)
```

```python
# use matplotlib syntax to put image in a figure

fig, ax = plt.subplots(figsize=(7,3))
ax.imshow(wordcloud,            #remember "imshow" from when we added pictures to matplotlib axes
          interpolation="bilinear") # this line helps smooth our image
ax.axis("off")
plt.show()
```

# venn diagrams

```python
# note: already imported venn diagram extension above
```

```python
# defining sets
A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])
```

```python
# assigning sets to each circle and modifying appearance

diagram = venn2_unweighted([A, B],
                           set_labels = ("Fruits", "Pies"),
                           set_colors = ("blue", "red"),
                           alpha=0.5)

plt.show()
# by default, output will only show counts of how many are in each set
```

```python
# modifying venn diagrams

diagram = venn2_unweighted([A, B],
                           set_labels = ("Fruits", "Pies"),
                           set_colors = ("blue", "red"),
                           alpha=0.5)

diagram.get_label_by_id("10").set_text("\n".join(A - B))
diagram.get_label_by_id("11").set_text("\n".join(A & B))
diagram.get_label_by_id("01").set_text("\n".join(B - A))

# numbers in brackets come from the documentation
# need to include backslash \n so words aren't all in a single line

plt.show()
```

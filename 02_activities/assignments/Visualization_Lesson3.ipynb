```python
# loading libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# sample data

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
# creating line plot

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1,y2)
```

```python
# adding a legend

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = "lower right")      # specifying the location of the legend within the box
```

```python
# modifying our legend -- add 4 different arguments

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label ="Person 2")
ax.legend(loc = "lower right",
          frameon = True,   # add frame around the legend
          fontsize = 12,    # change font size
          ncol = 2,         # specify number of columns
          shadow = True)     # adding drop shadow
```

```python
# moving legend outside of plot area -- adding bbox_to_anchor

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = "upper left",      # this is specifying the anchor point of the LEGEND,
                                    # so "upper right" would cause the legend to be within the box
          bbox_to_anchor = (1,1))
```

```python
# creating a scatter plot

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = "lower right")
```

```python
# using text() function to specify coordinates and text

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = "lower right")

ax.text(10,95, "This value is important")
```

```python
# modifying text annotations

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = "lower right")
ax.text(10,95, "This value is important",
        ha = "center",  # alignment of text
        color = "red",  # modify font color
        size = 20)   #modify font size
```

```python
# creating empty set of axes

fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
```

```python
# positioning annotation with reference to data (like above), the axes or the figure
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

ax.text (1,5, ". Data:(1,5)", transform=ax.transData)
ax.text(0.5, 0.1, ". Axes:(0.5,0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure)
    # the periods . are just to demonstrate where the annotation starts
```

```python
# annotating with arrows

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")

ax.annotate ("This is important!", xy=(10,95), xytext=(20,94),
             arrowprops=dict(facecolor="black", shrink=0.05))
```

```python
# modifying arrows with arrowprops argument
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")

ax.annotate("This is important!",
            xy=(10, 95), xytext=(20,94),
            arrowprops = dict(arrowstyle = "wedge",
                              color = "lightblue"))
```

```python
# removing tick marks and labels -- use locator to remove ticks and labels, use formatter to remove only labels

fig,ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# limiting tick marks

fig,ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label ="Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = "lower right")
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# setting tick marks at particular intervals

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# rotating axis labels and horizontal alignment -- can be used if x-axis tick marks are longer words
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")

plt.xticks(rotation=45, ha = "right")
```

```python
# modifying axis title

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")

fontx = {"family": "serif", "color":"indigo"}

plt.xlabel("Shiny New x Axis!", fontsize = 18, fontdict=fontx)
```
